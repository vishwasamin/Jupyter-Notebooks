## Filtering Out Data in Dataframes

```python
import pandas as pd
```

### 1) usecols

```python
data = pd.read_csv('employee_database.csv', usecols= ['emp_no','first_name','last_name','gender'])
data.head()
```

### 2) isin()

```python
data[data['gender'].isin(['M'])].head()
```

### 3) str.startswith

```python
data[data['first_name'].str.startswith('G')].head()
```

### 4) str.contains

```python
data[data['first_name'].str.contains('san')].head()
```

### 5) Only alphabets in one column

```python
passdata = pd.read_csv('data.csv', usecols=['password','category','time_unit'])
passdata.dropna(inplace=True)
passdata.head()
```

```python
passdata[passdata['password'].str.contains(r'^[A-Za-z]+$')]
```

### 6) Only numbers in column

```python
passdata[passdata['password'].str.contains(r'^[^A-Za-z]+$')].head()

# or
passdata[passdata['password'].str.contains(r'^[0-9]+$')].head()
```
